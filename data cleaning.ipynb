```python
def format_date(x):
    x = str(x)  # Convert to string in case it's not
    digits = re.sub(r"\s+", "", x)  # Remove all spaces
    match = re.match(r"(\d{4})(\d{2})(\d{2})", digits) # match dates for instance 20230412 is splitted by the first 4 digits, followed by the next two digits then the last 
    if match:
        return "/".join(match.groups()) # afterwars they are seperated uing a slash
    return x  # If it doesn't match expected format, return as-is
df["Date_of_Admission"] = df["Date_of_Admission"].apply(format_date)
df
```

```python
import re
from calendar import month_name

def clean_date_with_year(text):
    # Match "Early|Mid|Late" + 2-digit month + optional slash + 4-digit year
    match = re.search(r'\b(Early|Mid|Late)\s+(\d{2})[\/\s]*\s*(\d{4})', str(text))
    if match:
        position = match.group(1)
        month_num = int(match.group(2)) # group 2 is the maonth number gotten from the calendar module
        year = match.group(3) # group 3 is the year as indicated in the regex pattern
        return f"{position} {month_name[month_num]} {year}" # output  of the function
    return text

# Apply to your DataFrame column
df["Follow_up_Visit"] = df["Follow_up_Visit"].apply(clean_date_with_year)
```
